## Trump Fact Checking
Raw data unavailable, so assume poisson (not an a fully accurate assumption).
<br>https://www.washingtonpost.com/graphics/politics/trump-claims-database/?utm_term=.ee49ae958e3e

```python
import numpy as np
import scipy.stats
import collections
import matplotlib.pyplot as plt
import scipy.optimize
```

```python
num_days_of_history = 828
mu = 10111 / num_days_of_history # As of 5/24/19
str(int(np.round(mu))) + " lies per day"
```

```python
max_lies_per_day_to_analyze = 30
freq_expected = [scipy.stats.poisson.pmf(k = i, mu = mu) * num_days_of_history
                 for i in range(max_lies_per_day_to_analyze)]
```

```python
plt.plot(np.arange(max_lies_per_day_to_analyze), freq_expected, 
        linestyle = '-', color = 'black',
        label = 'Poisson')
plt.xlabel('Non-Facts per Day')
plt.ylabel('Number of Days With that Number of Non-Facts')
plt.title('Distribution Density of Non-Facts per Year')
plt.legend(loc='upper right')
plt.show()
```

## Arrival times

Chance of 1 no-fact within the next hour:

```python
time_to_investigate = 1/24 # number hours per day

prob_of_time_between_occurences = 1 - np.exp(-1 * mu * time_to_investigate) # this amount of time or less

str(int(np.round(prob_of_time_between_occurences * 100))) + ("% chance of a non-fact in the next hour.")
```
